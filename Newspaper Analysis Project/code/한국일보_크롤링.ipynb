```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re


# 뉴스 기사 내용을 저장할 리스트
news_contents = []

# 각 URL에 대해 HTML 내용을 크롤링
for i, url in enumerate(df['URL']):
    try:
        print(f"{i+1}: {url}")
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')

        # 해당하는 기사 내용 추출 (예시는 'div.tx'라고 가정)
        news_texts = soup.select('p.editor-p')

        # 각 기사 내용을 리스트에 추가
        news_content = ''
        for news_text in news_texts:
            clean_text = re.sub(r'[\r\n\t]', '', news_text.get_text(strip=True))
            news_content += clean_text + ' '

        if news_content:
            news_contents.append(news_content.strip())  # 문자열 양 끝의 공백 제거 후 리스트에 추가
        else:
            news_contents.append('내용 없음')  # 내용이 없는 경우 대체 텍스트


    except Exception as e:
        print(f"Error fetching {url}: {str(e)}")
        news_contents.append('크롤링 실패')  # 오류 처리

    # 서버 부하 방지를 위한 딜레이
    time.sleep(1)

# 원래 DataFrame에 기사 내용 칼럼 추가
df['기사내용'] = news_contents
```
