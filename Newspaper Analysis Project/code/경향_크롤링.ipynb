```python
# 엑셀 파일 로드
raw_df = pd.read_excel("경향신문.xlsx")

# 필요한 칼럼만 선택
raw_df = raw_df[['일자', '언론사', '제목', 'URL']]
```

```python
raw_df
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re

# 뉴스 기사 내용을 저장할 리스트
news_contents = []

# 각 URL에 대해 HTML 내용을 크롤링
for i, url in enumerate(raw_df['URL']):
    try:
        print(f"{i+1}: {url}")
        response = requests.get(url, headers={"User-Agent": "Mozilla/5.0"})
        html_content = response.text
        soup = BeautifulSoup(html_content, 'html.parser')
        
        # 해당하는 기사 내용 추출
        news_text = soup.select('p.content_text')
        
        # 텍스트 추출 및 공백 제거
        clean_text = ''
        for paragraph in news_text:
            clean_text += re.sub(r'[\r\n\t]', '', paragraph.get_text(strip=True)) + ' '

        # 리스트에 기사 내용 추가
        news_contents.append(clean_text.strip())

    except Exception as e:
        print(f"Error fetching {url}: {str(e)}")
        news_contents.append('크롤링 실패')  # 오류 처리

# 원래 DataFrame에 기사 내용 칼럼 추가
raw_df['기사내용'] = news_contents
```
